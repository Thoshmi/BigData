*String Compression Algorithm*

```python
def compress_message(msg):
    compressed_msg = ""
    count = 1

    for i in range(1, len(msg)):
        if msg[i] == msg[i - 1]:
            count += 1
        else:
            compressed_msg += msg[i - 1] + (str(count) if count > 1 else "")
            count = 1

    compressed_msg += msg[-1] + (str(count) if count > 1 else "")

    return compressed_msg
```

```python
# Input
msg = input()
# Output
compressed_msg = compress_message(msg)
print(compressed_msg)
```

*Web Scraping Faculty Information from Engineering and Computer Science Departments*

```python
import requests
from bs4 import BeautifulSoup

# URL of the faculty directory listing page
url = 'https://www2.eecs.berkeley.edu/Faculty/Lists/CS/faculty.html'

# HTTP GET request to the URL
response = requests.get(url)
```

```python
response
```

```python
 soup = BeautifulSoup(response.text, 'html.parser')
```

```python
faculty = soup.select('.media-heading')
```

```python
faculty
```

```python
prof_names = [x.get_text().rstrip() for x in faculty]
prof_names
```

```python
faculty_urls = []

for url in faculty:
  x_scrap = BeautifulSoup(str(url), 'html.parser')
  complete_url = "https://www2.eecs.berkeley.edu/"+x_scrap.find('h3').find('a')['href']
  faculty_urls.append(complete_url)
```

```python
faculty_urls
```

```python
faculty_profile = soup.select('.media-body')
```

```python
faculty_names = []
for profile in faculty_profile:
  profile_soup = BeautifulSoup(str(profile),'html.parser')
  anchors = profile_soup.find_all('a')
  faculty_names.append([ x.get_text() for x in anchors])
```

 *Faculty names and research*

```python
faculty_names
```

```python
biography = []
for faculty_url in faculty_urls:
  bio = ""
  response = requests.get(faculty_url)
  biography_soup = BeautifulSoup(response.text,'html.parser')
  final_output = biography_soup.find_all("div",{"class":"section bio> <h2>Biography</h2>"})
  if len(final_output) == 0:
    bio = "no bio available on website"
    biography.append(bio)
  else:
    info = [r.text.strip() for r in final_output]
    biography.append(info[0])
```

```python
with open('faculty_names.txt', 'w', encoding='utf-8') as file:
  for element in faculty_names:
    name = element[0].rstrip()
    file.write(f"Name: {name}\n")

  print("Names written to file.")
```

```python
#Research Areas
with open('professor_research_areas.txt', 'w', encoding='utf-8') as file:
  for element in faculty_names:
    elements_to_print = element[1:]
    result_string = ",".join(elements_to_print)

    file.write(f"Prof {element[0].rstrip()} Research Areas: {result_string}\n")
  print("Bios written to file.")
```

```python
#URLs
with open('professor_urls.txt', 'w', encoding='utf-8') as file:
  for url in faculty_urls:
    file.write(f"{url}\n")
print("URLs written.")
```
